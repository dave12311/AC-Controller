```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
samples = 2400

log = np.loadtxt("C:/Users/horva/Desktop/LA/01_new_new_3.txt")
bits = log[1:samples+1]
start = log[0]
end = log[samples+1]

for i in range(samples):
    if bits[i] == 1.0:
        bits[i] = 0
    elif bits[i] == 0.0:
        bits[i] = 1
        
length = end-start
step = length / samples

us = []
for i in range(samples):
    us.append(i*step)

plt.figure(figsize=(100,3),dpi=300)
plt.step(us,bits,where='mid')
plt.ylim(-0.1,1.1)
plt.xlim(us[0], us[-1])
plt.plot(us, bits, '.')
plt.show()
```
