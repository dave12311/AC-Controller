```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot(filename):
    log = np.loadtxt(filename, delimiter=',')
    us = log[:,0]
    bits = log[:,1]
    plt.figure(figsize=(40,3),dpi=500)
    plt.step(us, bits, where='mid')
    plt.ylim(-0.1, 1.1)
    plt.xlim(us[0], us[-1])
    plt.plot(us, bits, '.')
    plt.show()
```
